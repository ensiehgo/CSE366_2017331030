```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0.0,2.0,0.00001)
```

$$x(t) = A sin (2\pi ft+phase)$$

```python
def create_sinusoid(t,A,f,phase=0):
   x = A * np.sin(2.0 * np.pi * f * t + phase)
   return x
```

```python
x1 = create_sinusoid(t,1,5)
x2 = create_sinusoid(t,1,10)
x3 = create_sinusoid(t,1,15)
y = x1+x2+x3
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(11,11))
fig.suptitle('3 Sinusoidal Signals and 1 Composite Signal', fontsize=15)

# plotting signal-1
ax[0,0].plot(t,x1)
ax[0,0].set_title('Signal-1', fontsize=15)

# ploting signal-2
ax[0,1].plot(t,x2)
ax[0,1].set_title('Signal-2', fontsize=15)

# ploting signal-3
ax[1,0].plot(t,x3)
ax[1,0].set_title('Signal-3', fontsize=15)

# ploting signal-4
ax[1,1].plot(t,y)
ax[1,1].set_title('Signal-4', fontsize=15)
```

```python
fs = 10
step = 1.0/fs
n = np.arange(0.0, 2.0, step)
y = create_sinusoid(n,1,5)+create_sinusoid(n,1,10)+create_sinusoid(n,1,15)
plt.title('fs=10',fontsize=15)
plt.stem(n,y)
plt.show()

fs = 20
step = 1.0/fs
n = np.arange(0.0, 2.0, step)
y = create_sinusoid(n,1,5)+create_sinusoid(n,1,10)+create_sinusoid(n,1,15)
plt.title('fs=20',fontsize=15)
plt.stem(n,y)
plt.show()

fs = 50
step = 1.0/fs
n = np.arange(0.0, 2.0, step)
y = create_sinusoid(n,1,5)+create_sinusoid(n,1,10)+create_sinusoid(n,1,15)
plt.title('fs=50',fontsize=15)
plt.stem(n,y)
plt.show()
```
