```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,2,80)
```


  $$ x(t) =  A sin(2\pi ft+phase) $$


```python
def create_sinusoid(A, f, phase = 0):
  x = A * np.sin(2.0 * np.pi * f * t + phase)
  return x 
```

**Creating a Sinusoidal Signal $x(t)$ and 3 different plotting in subplots**

---



```python
x = create_sinusoid(1,1)

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(11,11))
fig.suptitle('Different subplots of a sinusoidal signal', fontsize=15)

ax[0,0].plot(t,x)
ax[0,0].set_title('continuous plot', fontsize=15)

ax[0,1].stem(t,x)
ax[0,1].set_title('stem plot', fontsize=15)

ax[1,0].scatter(t,x)
ax[1,0].set_title('scatter plot', fontsize=15)
ax[1,1].axis('off')

plt.show()
```

**Create three Sinusoidal Signal $x1(t), x2(t), x3(t)$ then comine and plot the signal**




```python
x1 = create_sinusoid(1,1)
x2 = create_sinusoid(2,5)
x3 = create_sinusoid(3,3)
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2, figsize=(11,11))
fig.suptitle('3 Sinusoidal signal and a Combined signal', fontsize=15)

ax[0,0].plot(t,x1,'r')
ax[0,0].set_title('x1(t)', fontsize=15)

ax[0,1].plot(t,x2,'g')
ax[0,1].set_title('x2(t)', fontsize=15)

ax[1,0].plot(t,x3,'b')
ax[1,0].set_title('x3(t)', fontsize=15)

ax[1,1].plot(t,x1+x2+x3)
ax[1,1].set_title('y(t) = x1(t)+ x2(t) + x3(t)', fontsize=15)
```
