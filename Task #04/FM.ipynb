```python
import numpy as np
import matplotlib.pyplot as plt
```

*Input*

```python
fm = float(input('Message Frequency = ')) # 1
fc = float(input('Carrier Frequency = ')) # 25
mi = float(input('Modulation Index = ')) # 10
t = np.arange(0,1,0.001)
```

**Constructing Modulated Signal from message and carrier signal**

```python
m = np.sin(2 * np.pi * fm * t) # Message Signal
c = np.sin(2 * np.pi * fc * t) # Carrier Signal
y = np.sin(2 * np.pi * fc * t + (mi * np.sin(2 * np.pi * fm * t))) # Modulated Signal
```

```python
'''
Message Signal
'''
plt.plot(t,m)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Message Signal')
plt.grid(True)
plt.show()

'''
Carrier Signal
'''
plt.plot(t,c)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Carrier Signal')
plt.grid('True')
plt.show()

'''
Modulated Signal
'''
plt.plot(t,y)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('FM Signal')
plt.grid(True)
plt.show()
```
