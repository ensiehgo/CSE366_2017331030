```python
import numpy as np
import matplotlib.pyplot as plt
```

Input

```python
Ac = 1
fc = int(input('enter carrier frequency'))
fm = int(input('enter message frequency')) # fm < fc
m = float(input('enter modulation index'))
Am = m * Ac
```

**Constucting Modulated Signal from Message and Carrier Signal**

```python
t1 = np.arange(0,1,0.001)
ym = np.sin(2*np.pi*fm*t1) # message signal
yc = np.sin(2*np.pi*fc*t1) # carrier signal
eq = (1 + m * ym) * (Ac * yc)
```

**Plotting**

```python
'''
Message Signal
'''
plt.plot(t1,ym)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Message Signal')
plt.grid(True)
plt.show()

'''
Carrier Signal
'''
plt.plot(t1,yc)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Message Signal')
plt.grid(True)
plt.show()

'''
Modulated Signal
'''
plt.plot(t1,eq)
plt.plot(t1,eq, 'r')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Modulated Signal')
plt.grid(True)
plt.show()
```
