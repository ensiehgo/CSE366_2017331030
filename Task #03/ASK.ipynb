```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# v = [0,1,1,0,1,1,0,0,1,1,0]
v = [1,0,1,1,0,1,0,1,1,1]
```

```python
dim = 100
vx = []
```

```python
for i in range(len(v)):
    f = np.ones(dim)
    x = f*v[i]
    vx = np.concatenate((vx,x))
```

```python
plt.subplot(3,1,1)
plt.plot(vx)
plt.show()
```

**Carrier Signal**

```python
plt.subplot(3,1,2)
t = np.linspace(0,2,len(vx))
f = 5
y = np.cos(2*np.pi*f*2*t)
plt.plot(t,y)
plt.show()
```

**Modulated Signal**

```python
plt.subplot(3,1,3)
mult = vx*y
plt.plot(t,mult)
plt.show()
```
