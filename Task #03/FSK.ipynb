```python
import numpy as np
import matplotlib.pyplot as plt
```

**Digital Signal**

```python
# v = [0,1,1,0,1,1,0,0,1,1,0]
v = [1,0,1,1,0,1,0,1,1,1]
```

```python
dim = 100
vx = []
```

```python
for i in range(len(v)):
    f = np.ones(dim)
    x = f*v[i]
    vx = np.concatenate((vx,x))
```

```python
plt.subplot(1,1,1)
plt.plot(vx)
```

**Carrier Signal**

```python
t = np.linspace(0,5,len(vx))
f1 = 2
f2 = 5
y1 = np.cos(2*np.pi*f1*2*t)
y2 = np.cos(2*np.pi*f2*2*t)
```

```python
plt.subplot(2,1,1)
plt.plot(t,y1)
plt.subplot(2,1,2)
plt.plot(t,y2)
```

**Modulated Signal**

```python
mi = []
for i in range(0,len(vx)):
    if vx[i] == 0:
       zero = np.array([y1[i]])
       mi = np.concatenate((mi,zero))
    else:
        one = np.array([y2[i]])
        mi = np.concatenate((mi,one))
```

```python
plt.subplot(1,1,1)
plt.plot(t,mi)
```
