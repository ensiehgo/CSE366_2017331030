```python
import numpy as np
import matplotlib.pyplot as plt
```

**Digital Signal**

```python
# v = [0,1,0,0,1,1,1,0,1,0,1]
v = [1,0,1,1,0,1,0,1,1,1]
np.shape(v)
dim = 100
vx = []
```

```python
for i in range(len(v)):
    f = np.ones(dim)
    x = f*v[i]
    vx = np.concatenate((x,vx))
```

```python
plt.subplot(4,1,1)
plt.plot(vx)
```

**Carrier Signals**

```python
t = np.linspace(0,5,len(vx))
f = 2
y1 = np.cos(2*np.pi*f*2*t)
y2 = -np.cos(2*np.pi*f*2*t)
```

```python
plt.subplot(4,1,2)
plt.plot(t,y1)
plt.subplot(4,1,3)
plt.plot(t,y2)
```

**Modulated Signal**

```python
m = []
for i in range(0,len(vx)):
    if vx[i] == 0:
       zero = np.array([y2[i]])
       m = np.concatenate((m,zero))
    else:
        one = np.array([y1[i]])
        m = np.concatenate((m,one))
```

```python
plt.subplot(4,1,4)
plt.plot(t,m)
```
